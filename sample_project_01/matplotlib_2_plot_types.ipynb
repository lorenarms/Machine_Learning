# Plot Types
### Types
* Line
* Bar
* Scatter
* Histogram
* Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
# Create some data

# 100 different vallues spanning the space
# from 0 to 10
x = np.linspace(0, 10, 100) 
x[:15]

```

## Line Plot (`plot()`)

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

## Scatter Plot (`scatter()`)

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# another scatter
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

## Make a plot from a dictionary

```python
prices = {"Single": 10,
         "Double": 18,
         "Triple": 26}
fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values())
ax.set(title="Prices", ylabel="Price ($)", xlabel="Size");
```

## Horizonatal Bar Graph

```python
fig, ax=plt.subplots()
ax.barh(list(prices.keys()), list(prices.values()))
ax.set(title="Prices", xlabel="Price ($)", ylabel="Size")
```

## Histogram

```python
# make data; standard distribution array
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots
### Option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5));

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(prices.keys(), prices.values());
ax4.hist(np.random.randn(100));
```

### Option 2

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(prices.keys(), prices.values());
ax[1,1].hist(np.random.randn(100));
```
