# Matplotlib Introduction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Basic Plot Creation

```python
plt.plot([1, 2, 3, 4], [11, 22, 33, 44]);
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y);
```

```python
fig = plt.figure() # creates figure
ax = fig.add_subplot() # adds some axes
ax.plot(x, y)
plt.show()
```

```python
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# recommended method
fig, ax = plt.subplots(1, 1)
ax.plot(x, y) # add data
```

## Example workflow

```python
# 0. import matplotlib and get it ready for plotting
import matplotlib.pyplot as plt

# 1. prepare the data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot",
      xlabel="x axis",
      ylabel="y axis")

# 5. Export and show
fig.savefig("sample-plot.png")
```
