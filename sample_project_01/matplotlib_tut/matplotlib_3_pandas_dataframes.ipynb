### Convert Price column to float

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,]','',regex=True).astype(float)
```

```python
car_sales
```

### Add a column for date of sale

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
```

```python
car_sales
```

### Add a column for Total Sales

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

### Plot the data 

```python
car_sales.plot(x="Sale Date", y="Total Sales")
```

### pyplot versus matplotlib object-oriented method?

* pyplot is good for quick visualizations
* matplotlib is more advanced


```python
over_50 = heart[heart["age"] > 50]
len(over_50)
```

```python
# Pyplot
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# change limit of the x-axis; 'ax' gives us access to the axis'
# ax.set_xlim([45,100]);

```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# customize
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

# Building a Plot from scratch

```python
# set the style
plt.style.use('seaborn-v0_8-darkgrid')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# data axes 0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # change the color scheme, check docs for more

# customize
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

# set the x limit to remove weird extra graph parts
ax0.set_xlim([50,80])

# legend
ax0.legend(*scatter.legend_elements(), title="Target");

# horizontal line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


# data axes 1

plt.style.use('seaborn-v0_8-darkgrid')
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="spring")

# customize
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Maximum Heart Rate");

ax1.set_ylim([60,200])

# legend
ax1.legend(*scatter.legend_elements(), title="Target");

# line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# figure title
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
